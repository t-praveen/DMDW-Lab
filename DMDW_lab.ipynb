```
import pandas as  pd
link='https://raw.githubusercontent.com/Anasuya-Sahoo/DMDW-Lab/main/Toyota.csv'
df=pd.read_csv(link)
```

```
df.isnull().sum()
```

```
df.dropna(inplace=True)
```

```
df.isnull().sum()
```

```
df.head()
```

```
import matplotlib.pyplot as plt
plt.scatter(df['Age'],df['Price'],c='green')
plt.xlabel('Age(Months)')
plt.ylabel('Price(rupees)')
plt.show()
```

```
plt.hist(df['KM'])
```

```
plt.hist(df['KM'],color='green',edgecolor='white',bins=5)
plt.title('Histogram for KM')
plt.xlabel('Kilometer')
plt.ylabel('frequency')
plt.show()
```

```
import numpy as np
counts=[979,120,12]
fueltype=('Petrol','Diesel','CNG')
index=np.arange(len(fueltype))
plt.bar(index,counts,color=['red','green','blue'])
plt.title('Bar-plot')
plt.xlabel('fueltypes')
plt.ylabel('frequency')
plt.xticks(index,fueltype,rotation=90)
plt.show()
```

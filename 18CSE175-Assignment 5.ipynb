```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
path="https://raw.githubusercontent.com/Anasuya-Sahoo/DMDW-Lab/main/student-mat.csv"
df=pd.read_csv(path)
df.head()
```

```python
dfs=df[['schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']]
dfs.head()
```

```python
dfs=dfs.replace('no',0)
dfs=dfs.replace(to_replace='yes',value=1)
dfs.head()
```

```python
n=np.array(dfs[['schoolsup','famsup']])
n=n.reshape(-1,2)
n.shape
```

```python
m=np.array(dfs[['romantic','internet']])
m=m.reshape(-1,2)
m.shape
```

```python
from scipy.spatial import distance
```

```python
dist_matrix=distance.cdist(n,m)
dist_matrix.shape
```

```python
print(dist_matrix)
```

```python
sb.heatmap(dist_matrix)
plt.show()
```
